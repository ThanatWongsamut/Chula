## Which date that has the minimum average number of comments per VDO?

```python
from datetime import datetime, timedelta

# Define the start date
date = datetime.strptime('2017-11-14', '%Y-%m-%d').date()

# Infinite while loop
while True:
    df_on_date = df[df['trending_dt'].dt.date == date]

    if (df_on_date.shape[0] != df_on_date['video_id'].nunique()):
        print(date, df_on_date.shape[0], df_on_date['video_id'].nunique())
        break
    if (df_on_date.shape[0] == 0):
        print("No duplicate video_id on each day")
        break
    # Increment the date by one day for the next iteration
    date += timedelta(days=1)
```

```python
data=df.groupby('trending_dt')['comment_count'].mean()
min_dates = data[data == data.min()].index.to_list()
min_dates_str = [min_date.strftime('%Y-%m-%d') for min_date in min_dates]

print("Answer:", min_dates_str)
```

## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
import json
with open('data/US_category_id.json') as fd:
    cat = json.load(fd)

cat_list = []
for d in cat['items']:
    cat_list.append((int(d['id']), d['snippet']['title']))

cat_df = pd.DataFrame(cat_list, columns=['id', 'category'])
```

```python
df_withcat = df.merge(cat_df, left_on='category_id', right_on='id')
```

```python
cat_each_dt = df_withcat.groupby(['trending_dt', 'category']).agg(count = ('views','sum')).reset_index()

sport_count = cat_each_dt[cat_each_dt['category'] == 'Sports'].rename(columns={
    'count':'sport_count'
})

comedy_count = cat_each_dt[cat_each_dt['category'] == 'Comedy'].rename(columns={
    'count':'comedy_count'
})

res = sport_count.merge(comedy_count, on='trending_dt')
print("Answer:", res[res['sport_count'] > res['comedy_count']]['trending_dt'].nunique())
```
