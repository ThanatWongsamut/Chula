## How many VDO's that contain at least one record (row) of "dislikes" more than "likes"?  <font color=red>DO NOT group by the data and make sure that you count only unique title!</font>

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
# making condition
condition = (df['trending_date'] == '18.22.01') & (df['comment_count'] > 10000)

print(f"Ans: Number of video that are trending on 22 Jan 2018 (18.22.01) with comments more than 10,000 is {df[condition].shape[0]} videos")
```

## Which date that has the minimum average number of comments per VDO?

```python
# check if any duplicate video_id in each day appear (in 1 day, 1 video should be count only 1 time)

from datetime import datetime, timedelta

# Define the start date
current_date = datetime.strptime('2017-11-14', '%Y-%m-%d').date()
current_passed = 0


while True:
    current_date_df = df[df['trending_date'] == current_date.strftime("%y.%d.%m")]

    if current_date_df.shape[0] != current_date_df['video_id'].nunique():
        print(f'On {current_date.strftime("%y.%d.%m")}, there are duplicate video_id')
        break

    if current_date_df.shape[0] == 0:
        current_date += timedelta(days=1)
        continue

    current_passed += current_date_df.shape[0]

    if current_passed == df.shape[0]:
        print("No duplicate video_id on each day")
        break

    current_date += timedelta(days=1)
```

#### On 18.16.05, there are duplicate video_id and maybe happen on other day too.

```python
# find average comment_count for each day 
mean_each_day = df.groupby('trending_date')['comment_count'].mean()

# get index (date) of the row that has the minimum average number of comments
min_days = mean_each_day[mean_each_day == mean_each_day.min()].index.to_list()

# convert %y.%d.%m to %d %B %Y
from datetime import datetime
min_days_str = [datetime.strptime(min_day, "%y.%d.%m").strftime("%d %B %Y") for min_day in min_days]

# display all min days seperate with ','
print(", ".join(min_days_str))
```

## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
# import catagory dataframe
import json
with open('data/GB_category_id.json') as f:
    cat = json.load(f)
cat_list = []
for d in cat['items']:
    cat_list.append((int(d['id']), d['snippet']['title']))
cat_df = pd.DataFrame(cat_list, columns=['id', 'category'])

# merge categorie dataframe to main dataframe with id
df_withcat = df.merge(cat_df, left_on='category_id', right_on='id')

# extract only category and views of Sports and Comedy
df_withcat['sport_view'] = df_withcat['views'] * (df_withcat['category'] == 'Sports')
df_withcat['comedy_view'] = df_withcat['views'] * (df_withcat['category'] == 'Comedy')

# sum view on each day
result = df_withcat.groupby('trending_date').agg(sport_sumview=('sport_view', 'sum'), comedy_sumview=('comedy_view', 'sum'))

condition = result['sport_sumview'] > result['comedy_sumview']

print(f"Ans: The number of days that there are more total daily views of VDO in 'Sports' than in 'Comady' is {result[condition].index.nunique()} days")
```
