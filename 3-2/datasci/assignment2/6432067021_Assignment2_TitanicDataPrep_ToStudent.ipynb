# 1) Load data & review the data

<font color='blue'>Q1) What is the shape of this dataset? (rows & columns)</font>

```python
# Get the number of rows and columns
num_rows, num_columns = df.shape

# Display the result
print(f"The DataFrame has {num_rows} rows and {num_columns} columns.")
```

# 2) Drop unqualified variables

*   Drop variables with missing > 50%
*   Drop categorical variables with flat values > 70% (variables with the same value in the same column)

<font color='blue'>Q2) How many columns do we have left?</font>




```python
num_rows = df.shape[0]

df_drop = df.copy()

dropped_columns = []
for col in df_drop.columns:
    
    # Drop variables with missing > 50%
    if not (df_drop[col].notnull().sum() > 0.5 * num_rows):
        dropped_columns.append(col)
        print(f"{col} is dropped, because missing > 50%")
        continue

    # Drop categorical variables with flat values > 70%
    if df_drop[col].value_counts().max() > 0.7 * num_rows:
        dropped_columns.append(col)
        print(f"{col} is dropped, because flat values > 70%")
        continue

for col in dropped_columns:
    df_drop.drop(col, axis = 1, inplace = True)

df2 = df_drop.copy()
print(f'After drop unqualified variables, we have {df2.shape[1]} columns left.')
```

```python
df2
```

# 3) Remove all rows with missing target (the variable "Survived")

<font color='blue'>Q3) How many rows do we have left?</font>

```python
df2.dropna(subset = ['Survived'], inplace = True)
print(f'We have {df2.shape[0]} rows left.')
```

```python
df2
```

# 4) Handle outliers

For the variable “Fare”, replace outlier values with the boundary values


*   If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
*   If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)

<font color='blue'>Q4) What is the mean of “Fare” after replacing the outliers (round 2 decimal points)?</font>



```python
q1 = df2['Fare'].describe()['25%']
q3 = df2['Fare'].describe()['75%']
iqr = q3 - q1

# Remove the outliers
df3 = df2.copy()
df3['Fare'] = df3['Fare'].apply(lambda x : max(x, q1 - 1.5 * iqr))
df3['Fare'] = df3['Fare'].apply(lambda x : min(x, q3 + 1.5 * iqr))

mean_fare = df3.Fare.mean()
print(f'The mean of "Fare" after replacing the outliers is {round(mean_fare, 2)}')
df3
```

# 5) Impute missing value



*   Show the number of missing values in each variable
*   Impute missing values with mean & mode
*   Show the number of missing values again after missing value imputation

<font color='blue'>Q5) Which variable has the largest number of missing values?</font>



```python
categorical = ['Survived','Pclass','Sex','Embarked']

df4 = df3.copy()

print('Before missing value imputation:')
for col in df4.columns:
    print(f'There are {df4[col].isnull().sum()} missing values in {col}.')
    if col in categorical:
        # for categorical using mode
        mode = df4[col].mode()[0]
        df4[col].fillna(mode, inplace = True)
    elif df4[col].dtype == 'object':
        # for string that is not categorical we can't use mode
        continue
    else:
        # mean for others
        mean = df4[col].mean()
        df4[col].fillna(mean, inplace = True)

print('\nAfter missing value imputation:')
for col in df4.columns:
    print(f'There are {df4[col].isnull().sum()} missing values in {col}.')
df4
```

# 6) Convert categorical to numeric values



*   Drop the variables “Name” & “Ticket”
*   For the variables “Sex” & “Embarked”, perform the dummy coding and drop the first level. Also, drop those original variables (“Sex” & “Embarked”)

<font color='blue'>Q6) How many columns do we have?</font>



```python
# Drop the variables "Name" & "Ticket"
df4_drop = df4.drop(['Name', 'Ticket'], axis = 1)

# Dummy code for "Sex" & "Embarked" and drop the first level
df5 = pd.get_dummies(df4_drop, columns = ['Sex', 'Embarked'], drop_first = True)

print(f'We have {df5.shape[1]} columns left.')

df5
```

# 7) Partition data



*   Split train/test split with stratification using 70%:30% and random seed with 12345
*   Show a proportion between survived (1) and died (0) in all data sets (total data, train, test)

<font color='blue'>Q7) What is a proportion between survived (1) and died (0) in the training data?</font>



```python
from sklearn.model_selection import train_test_split

y = df5.Survived
X = df5.drop('Survived', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.3, random_state = 12345)

print(f'The proportion between survived (1) and died (0) in total data is {y.sum() / (len(y) - y.sum())}')
print(f'The proportion between survived (1) and died (0) in train data is {y_test.sum() / (len(y_test) - y_test.sum())}')
print(f'The proportion between survived (1) and died (0) in test data is {y_train.sum() / (len(y_train) - y_train.sum())}')

print(f'\nThere are {X_train.shape[0]} rows in training set.')
print(f'There are {X_test.shape[0]} rows in testing set.')
```
